```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/Protein_Supply_Quantity_Data.csv")
df.head()
```

```python
print(df.shape)

print(df.columns)

df.describe()
```

```python
print(df.isnull().sum())


df = df.dropna()

print("Cleaned! Remaining rows:", len(df))
```

```python
# Sort by Meat protein quantity in descending order and select the top 10 countries
top_10_meat = df[['Country', 'Meat']].sort_values(by='Meat', ascending=False).head(10)

print("Top 10 countries getting the most protein from Meat:")
print(top_10_meat.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation coefficient (r)
correlation = df['Animal Products'].corr(df['Obesity'])
print(f"Correlation coefficient (r) between Animal Products and Obesity: {correlation:.4f}")

# Plotting the scatter plot to visualize the trend
sns.scatterplot(data=df, x='Animal Products', y='Obesity')
plt.title('Relationship between Animal Products Protein & Obesity Rate')
plt.xlabel('Animal Products Protein ($Pct$)')
plt.ylabel('Obesity Rate ($Pct$)')
plt.savefig('animal_products_vs_obesity.png')
```

```python
import pandas as pd

# Convert Undernourished column to numeric format
df['Undernourished_numeric'] = pd.to_numeric(df['Undernourished'], errors='coerce')

# Sort by the numeric Undernourished column in descending order
top_10_undernourished = df[['Country', 'Undernourished_numeric']].sort_values(by='Undernourished_numeric', ascending=False).head(10)

print("Top 10 countries with the highest Undernourishment rate:")
print(top_10_undernourished.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean background style for the plot
sns.set_theme(style="ticks")

# Create a scatter plot with an integrated regression trendline
sns.regplot(
    data=df,
    x='Vegetal Products',
    y='Obesity',
    scatter_kws={'alpha': 0.6, 'color': '#2ca02c'},  # Green dots for plant/vegetal theme
    line_kws={'color': '#d62728', 'linewidth': 2}     # Red line to clearly show the trend direction
)

# Customize the chart title and labels
plt.title('The Impact of Plant-Based Diets: Vegetal Products Protein vs. National Obesity Rates', fontsize=12, pad=15)
plt.xlabel('Vegetal Products Protein Supply ($Pct$)', fontsize=10)
plt.ylabel('Obesity Rate ($Pct$)', fontsize=10)

# Automatically adjust layout to ensure labels are fully readable and non-overlapping
plt.tight_layout()

# Save the plot directly as a high-resolution image file
plt.savefig('vegetal_products_vs_obesity.png', dpi=300)
```
